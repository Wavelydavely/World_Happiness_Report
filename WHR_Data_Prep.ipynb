```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# Read data and store into a Pandas DataFrame
world_happiness_df = pd.read_csv("Resources/world-happiness-report.csv")
world_happiness_df.head()
```

```python
world_happiness_df = world_happiness_df.drop(["Positive_affect", "Negative_affect"], axis = 1)
world_happiness_df.head()
```

```python
world_happiness_df.rename(columns = {'Country_name': 'Country', 'year': 'Year', 'Life_Ladder': 'Happiness_Score', 'Log_GDP_per_capita': 'Economic_Production', 'Social_support': 'Social_Support', 'Healthy_life_expectancy_at_birth': 'Life_Expectancy', 'Freedom_to_make_life_choices': 'Freedom', 'Perceptions_of_corruption': 'Perceptions_of_Corruption'}, inplace = True)
world_happiness_df.head()                                                      
```

```python
world_happiness_df.dtypes 
```

```python
world_happiness_df.to_csv('Cleaned_Data/World_Happiness_Report_All_Years.csv', index=False)
```
